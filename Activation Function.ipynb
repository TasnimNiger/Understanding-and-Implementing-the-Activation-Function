```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Activation Function
def sigmoid(x):
  return 1/(1+np.exp(-x))

# Derivative of Sigmoid
def der_sigmoid(x):
  return sigmoid(x) * (1- sigmoid(x))

# Hyperbolic Tangent (htan) Activation Function
def tanh(x):
  return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

# htan derivative
def der_tanh(x):
  return 1 - tanh(x) * tanh(x)

# Rectified Linear Unit (ReLU)
def ReLU(x):
  data = [max(0,value) for value in x]
  return np.array(data, dtype=float)

# Derivative for ReLU
def der_ReLU(x):
  data = [1 if value>0 else 0 for value in x]
  return np.array(data, dtype=float)

# Leaky Rectified Linear Unit (leaky ReLU) Activation Function
def leaky_ReLU(x):
  data = [max(0.05*value,value) for value in x]
  return np.array(data, dtype=float)

# Derivative for leaky ReLU 
def der_leaky_ReLU(x):
  data = [1 if value>0 else 0.05 for value in x]
  return np.array(data, dtype=float)

def softmax(x):
 data = np.exp(x - np.max(x))/ np.sum(np.exp(x - np.max(x)), axis=0)
 return np.array(data, dtype=float) 

def binary_step(x):
  data = [1 if value>=0 else 0 for value in x]
  return np.array(data, dtype=float)

# Generating data to plot
x_data = np.linspace(-3,3,100)

y_sig_data = sigmoid(x_data)
dy_sig_data = der_sigmoid(x_data)

y_tanh_data = tanh(x_data)
dy_tanh_data = der_tanh(x_data)

y_relu_data = ReLU(x_data)
dy_relu_data = der_ReLU(x_data)

y_lrelu_data = leaky_ReLU(x_data)
dy_lrelu_data = der_leaky_ReLU(x_data)

y_soft_data =softmax(x_data)

y_bi_data =binary_step(x_data)

# Create subplots for each activation function
fig, axs = plt.subplots(3, 2, figsize=(15, 20))

# Plotting
axs[0, 0].plot(x_data, y_sig_data, x_data, dy_sig_data)
axs[0, 0].set_title('Sigmoid Activation Function & Derivative')
axs[0, 0].legend(['sigmoid','der_sigmoid'])
axs[0, 0].grid(True)

axs[0, 1].plot(x_data, y_tanh_data, x_data, dy_tanh_data)
axs[0, 1].set_title('tanh Activation Function & Derivative')
axs[0, 1].legend(['tanh','der_tanh'])
axs[0, 1].grid(True)


axs[1, 0].plot(x_data, y_relu_data, x_data, dy_relu_data)
axs[1, 0].set_title('ReLU Activation Function & Derivative')
axs[1, 0].legend(['ReLU','der_ReLU'])
axs[1, 0].grid(True)

axs[1, 1].plot(x_data, y_lrelu_data, x_data, dy_lrelu_data)
axs[1, 1].set_title('leaky ReLU Activation Function & Derivative')
axs[1, 1].legend(['leaky_ReLU','der_leaky_ReLU'])
axs[1, 1].grid(True)

axs[2, 0].plot(x_data, y_soft_data)
axs[2, 0].set_title('Softmax Activation Function')
axs[2, 0].legend(['softmax'])
axs[2, 0].grid(True)

axs[2, 1].plot(x_data, y_bi_data)
axs[2, 1].set_title('Binary step Activation Function')
axs[2, 1].legend(['Binary step'])
axs[2, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
